```python

#Funções
def menu_fun():
    print("############################")
    print("*     CALCULADORA          *")
    print("*                          *")
    print("* (1). Adição              *")
    print("* (2). Subtração           *")
    print("* (3). Multiplicação       *")
    print("* (4). Divisão             *")
    print("* (5). Exponenciação       *")
    print("* (6). Raiz quadrada       *")
    print("* (7). Porcentagem         *")
    print("* (8). Resto da divisão    *")
    print("* (9). Divisão de inteiro  *")
    print("* (0). Sair do programa    *")
    print("*                          *")
    print("############################")
    try:
        return float(input("Opção Menu: "))
    except:
        pass

def soma():
    num1 = float(input("Digite o primeiro número para a adição: "))
    num2 = float(input("Digite o segundo número para a adição: "))
    resultado_adicao= num1 +num2
    return resultado_adicao


def subtracao():
    num1 = float(input("Digite o primeiro número para a subtração: "))
    num2 = float(input("Digite o segundo número para a subtração: "))
    resultado_subtracao= num1 - num2
    return resultado_subtracao



def multiplicacao():
    num1 = float(input("Digite o primeiro número para a multiplicação: "))
    num2 = float(input("Digite o segundo número para a multiplicação: "))
    resultado_multiplicacao= num1 * num2
    return resultado_multiplicacao

def divisao():
    num1= float(input("Informe o numero para ser dividido: "))
    num2= float(input("Informe o divisor : "))
    if num2 != 0:
        resultado_divisao= num1 / num2
        return resultado_divisao
    else:
        return "Divisor precisa ser diferente de Zero "


def exponenciacao():
    num1 = float(input("Digite a base para a exponenciação: "))
    num2 = float(input("Digite o expoente para a exponenciação: "))
    resultado_exponenciacao= num1 ** num2
    return resultado_exponenciacao


def raiz_quadrada():
    num= float(input("Digite o número para calcular a raiz quadrada: "))
    resultado_raiz_quadrada= num **0.5
    return resultado_raiz_quadrada

def porcentagem():
    num1= float(input("Digite a porcentagem (sem o símbolo %): "))
    num2= float(input("Informe número para calcular a porcentagem: "))
    percentual= num1/100
    resultado_porcentagem= num2 * percentual
    return resultado_porcentagem


def resto_divisao():
    num1 = float(input("Digite o primeiro número para o cálculo do resto da divisão: "))
    num2 = float(input("Digite o segundo número para o cálculo do resto da divisão: "))
    resto_divisao= num1 % num2
    return resto_divisao

def divisao_inteiro():
    num1 = float(input("Digite o primeiro número para a divisão inteira: "))
    num2 = float(input("Digite o segundo número para a divisão inteira: "))
    divisao_inteiro= num1 // num2
    return divisao_inteiro

#Inicio
while True:
    menu= menu_fun()
    match menu:
        case 0:
            print("Você Saiu!")
            break
        case 1:
            resultado = soma()
            print(f"Resultado: {resultado}\n" + "*" * 27)
        case 2:
            resultado = subtracao()
            print(f"Resultado: {resultado}\n" + "*" * 27)
        case 3:
            resultado = multiplicacao()
            print(f"Resultado: {resultado}\n" + "*" * 27)
        case 4:
            resultado = divisao()
            print(f"Resultado: {resultado}\n" + "*" * 27)
        case 5:
            resultado = exponenciacao()
            print(f"Resultado: {resultado}\n" + "*" * 27)
        case 6:
            resultado = raiz_quadrada()
            print(f"Resultado: {resultado:.2f}\n" + "*" * 27)
        case 7:
            resultado = porcentagem()
            print(f"Resultado: {resultado}\n" + "*" * 27)
        case 8:
            resultado = resto_divisao()
            print(f"Resultado: {resultado}\n" + "*" * 27)
        case 9:
            resultado = divisao_inteiro()
            print(f"Resultado: {resultado}\n" + "*" * 27)
        case _:
            print("Opção inválida. Escolha uma opção entre 1 e 9 ou 0 para sair.")
```
